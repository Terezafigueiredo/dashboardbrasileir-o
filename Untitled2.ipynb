```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("Dashboard Brasileirão").getOrCreate()
df = spark.read.csv("dados_2018_2022.csv", header=True, inferSchema=True)
df.show(5)
#1 passo Se você já tem o CSV salvo, carregue com:
```

```python
#Criar coluna de resultado (Vitória, Empate, Derrota)
from pyspark.sql.functions import when

df = df.withColumn("resultado_mandante", when(df["gols_mandante"] > df["gols_visitante"], "V")
                   .when(df["gols_mandante"] == df["gols_visitante"], "E")
                   .otherwise("D"))
```

```python
#Validar os dados 
df.printSchema()
df.select("ano_campeonato").distinct().orderBy("ano_campeonato").show()
df.select("resultado_mandante").groupBy("resultado_mandante").count().show()
```

```python
#Total de jogos por time mandantes
jogos_por_time = df.groupBy("time_mandante").count().orderBy("count", ascending=False)
jogos_por_time.show()
```

```python
#Total de jogos por time mandantes
resultados = df.groupBy("time_mandante", "resultado_mandante").count().orderBy("time_mandante")
resultados.show()
```

```python
#Média de gols marcados e sofridos por time mandante
from pyspark.sql.functions import avg

medias_gols = df.groupBy("time_mandante") \
    .agg(avg("gols_mandante").alias("media_gols_marcados"),
         avg("gols_visitante").alias("media_gols_sofridos")) \
    .orderBy("media_gols_marcados", ascending=False)

medias_gols.show()
```

```python
#Evolução de desempenho por temporada Gols por ano:
gols_por_ano = df.groupBy("ano_campeonato") \
    .agg(avg("gols_mandante").alias("media_gols_mandante"),
         avg("gols_visitante").alias("media_gols_visitante")) \
    .orderBy("ano_campeonato")

gols_por_ano.show()
```

```python
#Vitórias por ano:
vitorias_por_ano = df.filter(df["resultado_mandante"] == "V") \
    .groupBy("ano_campeonato").count().orderBy("ano_campeonato")

vitorias_por_ano.show()
```
